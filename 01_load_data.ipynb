## Reformat images
A function to:
* Scale the images to a common size and data type 
* Transform images into numpy arrays
* Test the function by plotting the first image

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from skimage.transform import resize
```

```python
# all images will be resized to this
width = 179
height = 179

def format_image(path):
    """Takes a path to an image as input. Returns a numpy array."""
    # Load the image
    img = plt.imread(path)
    # resize the image
    img = resize(img, (height, width), anti_aliasing=True)
    return(img)
```

Plot the first image

```python
path = "data/barrel_jellyfish/01.jpeg"
img = format_image(path)

# print image properties
print('shape:', img.shape) # (179, 179, 3)
print('min:', img.min(), 'max:', img.max()) # min: 0 max: 1
print('dtype:', img.dtype) # float64

# Plot
plt.imshow(img)
plt.show()
```
